## NumPy
We use NumPy to create and manipulate list of numbers.  
We want to create points in order to show them in a graph.  
These are the x values of our points.

```
# import numpy and create array of 100 numbers from 0 to 99
import numpy as np
x = np.arange(100)
x
```

In order to create the y values of the points, we need to decide on a function.  
Let's use the indentity function `y(x) = x`.

```
y = x.copy()
y
```

## Matplotlib
We use Matplotlib.pyplot to show graphs.  
We can plot the dots in a scatter plot, to show the the function with the dots.  

```
# import matplotlib.pyplot, enlarge figure size and create scatter plot
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
plt.scatter(x, y)
```

We created and showed the function `y(x) = x`.  
The data in real life is hardly this plain and simple.  
Most of the time the data is messy or noisy.  
Let's try to add some noise (also called error) to the function points and create more natural looking dataset (a dataset is a set of data samples).  
We do it with random noise in [normal distribution](https://en.wikipedia.org/wiki/Normal_distribution), with [standard deviation](https://en.wikipedia.org/wiki/Standard_deviation) of 5.  

```
# create noise with normal distribution with mean=0 and standard_deviation=5
eps = np.random.normal(loc=0, scale=5, size=[len(y)])
eps
```

We call the natural looking data points `t` for *true values*. 

```
# add the noise to the function points' y values
t = y + eps
t
```

We can show both the new points with the old function together in the same graph.  

```
# show graph of our original function and the new data points
plt.scatter(x, t)
plt.plot(x, y, '+-r')
plt.show()
```

We want to find out how much the new points are far from the original function.  
In order to do so, we need to develop a way of evaluating this value.  
First let's show the distance of each point from it's original position.  

```
# add lines that represent the distance of each noisy point from the original point
plt.scatter(x, t)
plt.plot(x, y, '+-r')
for px, py, pt in zip(x, y, t):
    plt.plot([px, px], [py, pt], "-g")
plt.grid()
plt.show()
```

We can sum all the distances and get a number that represent the error of all the points, but then, with more points, we will have bigger error value (we sum more values).  
We want to get a value that will give the similar value range with less or more points.  
So, we can divide the summation by the number of points and get the mean error.  
The only thing we have to dicide on is the way we calculate the distance.  
This distance should be positive, no matter where is the noisy point in the graph (above or below the function line).  
So, we have 2 options:
1. Take the distance and raise it to the power of 2 to get `dist^2`. This approach is called **MSE - Mean Squared Error**.
2. Take the absolute value of the distance to get `|dist|`. This approace is called **MAE - Mean Absolute Error**.  

You can read more on the differences between them in [MAE and RMSE — Which Metric is Better?
](https://medium.com/human-in-a-machine-world/mae-and-rmse-which-metric-is-better-e60ac3bde13d) (RMSE is the squared root of the MSE, it is very simillar and for our usage their is no difference in using one over the other).  

Let's use the MSE approach.  


```
# calculate MSE of the points in the graph
sum_dist = 0
for px, py, pt in zip(x, y, t):
    sum_dist += (py - pt)**2
mean_dist = sum_dist/len(y)
mean_dist
```

We can define methods that will help us automate this process and create graphs for any function.  

```
# show graph of x, y and t values
def show_graph(x, y, t):
    plt.scatter(x, t)
    plt.plot(x, y, '+-r')
    for px, py, pt in zip(x, y, t):
        plt.plot([px, px], [py, pt], "-g")
    plt.grid()
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(f'MSE: {calculate_mse(y, t)}')

# calculate the mse for x, y and t values
def calculate_mse(y, t):
    sum_dist = 0
    for py, pt in zip(y, t):
        sum_dist += (py - pt)**2
    mse = sum_dist/len(y)
    return mse

show_graph(x, y, t)
```

We can also define a method that will create the new noisy points from the original points.  
We can use NumPy `sin` method to create the function `y(x) = sin(x)`. 

```
# create noisy y from original y
def create_t_from_y(y, sd):
    return y + np.random.normal(loc=0, scale=sd, size=[len(y)])

y = np.sin(x)
t = create_t_from_y(y, 5)
show_graph(x, y, t)
```

## ScyPy
We have a problem with this graph, the lines of the `sin` function are not rounded enough. It is just a bunch of red lines and it is not resembling much of the `sin` fucntion we know.  
We can use ScyPy interpolation to make the lines and their connections smoother.  

```
# import `interp1d` module form `scipy.iterpolate` library and show the graph with the interpolation
from scipy.interpolate import interp1d

def show_graph_with_interpolation(x, y, t, title):
    xnew = np.linspace(x.min(), x.max(), num=len(x)*10, endpoint=True)
    f3 = interp1d(x, y, kind='quadratic')
    plt.plot(xnew, f3(xnew), '-r')
    plt.scatter(x, t)
    plt.plot(x, y, '+r')
    for px, py, pt in zip(x, y, t):
        plt.plot([px, px], [py, pt], "-g")
    plt.grid()
    plt.xlabel('X', fontsize=20)
    plt.ylabel('Y', fontsize=20)
    plt.title(f'{title}, MSE: {calculate_mse(y, t)}', fontsize=25)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)

show_graph_with_interpolation(x, y, t, 'sin')
```

Let's define a method that will create any function we want (well.., we need to create the support for them in this method, but let's pretend we can create any function we want).

```
# showes the function with the number of points specified
# we can change the standard deviation of the noisy points
def plot_func_and_rand_points(x, func, sd, func_name='????'):
    y = func(x)
    t = create_t_from_y(y, sd)
    show_graph_with_interpolation(x, y, t, f'{func_name} fucntion, sd: {sd}')
```

Now, we can investigate the effect of the standard deviation on the MSE.  
Let's try to insert different `sd` for different functions and see how does the MSE changes.

```
# show sd=[5, 3, 1, 0] for linear function
plot_func_and_rand_points(np.arange(100), lambda x: x.copy(), 5, 'linear')
plt.show()
plot_func_and_rand_points(np.arange(100), lambda x: x.copy(), 3, 'linear')
plt.show()
plot_func_and_rand_points(np.arange(100), lambda x: x.copy(), 1, 'linear')
plt.show()
plot_func_and_rand_points(np.arange(100), lambda x: x.copy(), 0,' linear')
plt.show()
```

```
# show sd=[5, 3, 1, 0] for sin function
plot_func_and_rand_points(np.arange(100), np.sin, 5, 'sin')
plt.show()
plot_func_and_rand_points(np.arange(100), np.sin, 3, 'sin')
plt.show()
plot_func_and_rand_points(np.arange(100), np.sin, 1, 'sin')
plt.show()
plot_func_and_rand_points(np.arange(100), np.sin, 0, 'sin')
plt.show()
```

```
# show sd=[5, 3, 1, 0] for log function
plot_func_and_rand_points(np.arange(2, 102), np.log, 5, 'log')
plt.show()
plot_func_and_rand_points(np.arange(2, 102), np.log, 3, 'log')
plt.show()
plot_func_and_rand_points(np.arange(2, 102), np.log, 1, 'log')
plt.show()
plot_func_and_rand_points(np.arange(2, 102), np.log, 0, 'log')
plt.show()
```

We can see that the bigger the `sd`, the bigger the MSE.  
When the `sd` is `0`, the MSE is `0` too (all the noisy points are exactly the same as the original points - there is no noise\error).  
We can also check, what will happen if we will try to calculate the MSE between points of two different functions.  

```
# showes both functions with number of points specified and calculate MSE
# for log, we take points from 2 and not from 0 because log(0) is infinity and 
# log(1) is 0
def plot_func_and_rand_points_two_functions(x, func1, func2, func1_name='????', func2_name='????'):
    y1 = func1(x)        
    y2 = func2(x)        
    show_graph_with_interpolation(x, y1, y2, f'{func1_name} fucntion with points of {func2_name} function')
```

```
# show MSE of some combinations of functions 
plot_func_and_rand_points_two_functions(np.arange(100), np.sin, np.cos, 'sin', 'cos')
plt.show()
plot_func_and_rand_points_two_functions(np.arange(2, 102), lambda x: x.copy(), np.log, 'linear', 'log')
plt.show()
```

We can see that sometimes the MSE is big and somtimes it is not so bad.  
We can also see that even if in the sin-cos combination, the functions are totaly different, the MSE is low, because the values are low. There are no points that lies very far from the funciton line.  
In the linear-log combination, the points are very far from the function line. That makes the MSE huge.  

We can try to reduce the linear-log MSE, by changing the slope of the linear funtion.

```
# show MSE of some combinations of `linear` (m=[0.5, 0.05, 0.005]) and `log`
plot_func_and_rand_points_two_functions(np.arange(2, 102), lambda x: 0.5*x.copy(), np.log, 'linear m=0.5', 'log')
plt.show() 
plot_func_and_rand_points_two_functions(np.arange(2, 102), lambda x: 0.05*x.copy(), np.log, 'linear m=0.05', 'log')
plt.show()
plot_func_and_rand_points_two_functions(np.arange(2, 102), lambda x: 0.005*x.copy(), np.log, 'linear m=0.005', 'log')
plt.show()
```

We can write the slopes and the MSE to try to figure out what is the best slope that gives us the lowest MSE:  

| slope | 1    | 0.5 | 0.05 | 0.005 |
|-------|------|-----|------|-------|
| MSE   | 3075 | 673 | 1    | 12    |

We can see that at first the MSE is getting lower along with the slope, but this trend flips when we get to `slope=0.05` and `slope=0.005`.  
The MSE is suddenly getting larger.  
It means that the best MSE value for these functions, is lying somewhere between `slope=0.5` and `slope=0.005`.  The best slope in our attempts is `slope=1` but we can not know if it is the best one or maybe there are better slopes out there (a hint, somthing out there may look better, but is it worth the trouble?). 

## SymPy
Let's see the defenition of the MSE expression we created in matematics.  
To do this we use SymPy library.  
We need to build the expression like a lego from building blocks.  
We first define the smallest building blocks (symbols, indexes, functions, etc.).  
Then, we can build the expression from these blocks.  
The expressions are rendered with [Unicode](https://en.wikipedia.org/wiki/Unicode).    

```
# import all the moduls inside `sympy` library and create the basic building blocks for our expression
from sympy import *
n, i = symbols('n i')
xi = Indexed('x', i)
ti = Indexed('t', i)
yf = Function('y')
init_printing(use_unicode=True)
```

```
# create the MSE expression from the basic building blocks
MSE = (1/n)*Sum((ti-yf(xi))**2, (i, 1, n))
MSE
```

We can also differentiate the MSE expression with SymPy.

```
# get the derivative of MSE
MSE_dif = diff(MSE, xi)
MSE_dif
```

Let's see the string representation of the above expression.

```
# show the string representation of MSE expression in sympy
srepr(MSE)
```

We can also show the graph of this expression.  
It is a graph with nodes and edges, that is in a [tree form](https://en.wikipedia.org/wiki/Tree_(graph_theory)).  

```
# show the graph as a string with pprint (pretty printing)
import pprint
from sympy.printing.dot import dotprint

pprint.pprint(dotprint(MSE))
```

We want to see the graph in a more interpretable visualization.  
We can do it with `graphviz` library.  
We render the graph in [**SVG** format](https://he.wikipedia.org/wiki/SVG).  
This will save the graph image sharp in any size.  

```
# import `Source` module form `graphviz` library and use it to graph the tree
# save it in SVG and display it
from graphviz import Source
from IPython.display import SVG 
import os

def draw_tree(expr, delete_file=True, file_name='temp'):
    src = Source(dotprint(expr))
    src.render(file_name, format='svg')
    svg = SVG(filename=(f'{file_name}.svg'))
    os.unlink(f'{file_name}.svg')
    return svg

display(MSE)
draw_tree(MSE)
```

Let's calculate the MSE value from our points and funtions.  
In order to do so, we will define a method that represents a function that gets `x` and `t` (original points) and returns the noisy points.  
We will transfer this `y` funtion with the `x` and `t` variables to the `mse` method, to calculate the MSE from the SymPy expressoin.  

```
# get the `y_func` we want to evaluate in the MSE expression
def get_y_func(func_name, sd):
    def inner_func(x):
        SUPPORTED_FUNCTIONS = ['linear', 'sin', 'cos', 'exp']
        
        if func_name == 'linear':
            y = lambda x: x
        elif func_name == 'sin':
            y = np.sin
        elif func_name == 'cos':
            y = np.cos
        elif func_name == 'exp':
            y = np.exp
        else:
            raise ValueError(f'func_name variable must be one of {SUPPORTED_FUNCTIONS}.\n\
                you enterd `{func_name}`')
            
        eps = np.random.normal(scale=sd)
        return y(x) + eps
    
    return inner_func

# evaluate the MSE of the `y_func` on the data points (x, t)
def mse(x, t, y_func):
    n = symbols('n')
    xi = Indexed('x', i)
    ti = Indexed('t', i)
    yf = Function('y')
    MSE = (1/n)*Sum((ti-yf(xi))**2, (i, 1, n))
    one_on_n = MSE.args[0].subs({n: len(x)})
    sum_i_to_n = MSE.args[1].subs({n: len(x)})
    y = list(map(y_func, x))
    dist_pow_2 = (t - y)**2
    return MSE.func(one_on_n.doit(), Subs(sum_i_to_n.doit(), [sum_i_to_n.function.subs(sum_i_to_n.variables[0], j) for j in range(sum_i_to_n.limits[0][1], sum_i_to_n.limits[0][2] + 1)], dist_pow_2).doit()).doit()
```

Let's use the methods we created to calculate the MSE of 100 points that was created from the `sin` function (with noise).

```
# create the original points that will be used in MSE
x = np.arange(100)
t = np.sin(x)

# create the function to enter the mse
y_func = get_y_func('sin', 5)

# calculate the mse
mse(x, t, y_func)
```
